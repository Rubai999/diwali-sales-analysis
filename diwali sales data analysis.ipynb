```python
# EXPLORATORY DATA ANALYSIS
```

```python
# GENDER
```

```python
ax=sns.countplot(x='Gender',data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sns.countplot(x='Gender',data=df,label=True)
plt.show()
```

```python
sales_gen=df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Gender',y='Amount',data=sales_gen)
plt.show()
# here as_index=false is used to prevent gender from becoming index and keep it as a individual column
```

```python
se=df.groupby(['Gender'],as_index=False)['Amount'].sum()#.sort_values(by='Amount',ascending=False)
sns.barplot(x='Gender',y='Amount',data=se)

plt.show()
```

```python
df.groupby(['Gender'])['Amount'].sum()
```

```python
# from the above graphs it can  be noted that most of the buyers are females and they also have greater purchasing power than men 
```

```python
# AGE
```

```python
ax=sns.countplot(x='Age Group',data=df,hue='Gender')
for bars in ax.containers:
    ax.bar_label(bars)
 
```

```python

sales_age=df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Age Group',y='Amount',data=sales_age)
plt.show()
```

```python
# most of the buyers are female and between 26-35 yrs of age.
```

```python
df.columns
```

```python
# State
```

```python
sales_state=df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,10)})
sns.barplot(x='State',y='Orders',data=sales_state)
plt.show()
```

```python
sales_state=df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,10)})
sns.barplot(x='State',y='Amount',data=sales_state)
plt.show()
```

```python
# most of the orders are from uttar pradesh and their amount of spending is also greater 
```

```python
# Shaadi
```

```python
ax=sns.countplot(x='Shaadi',data=df)
sns.set(rc={'figure.figsize':(10,6)})

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
shadi_sales=df.groupby(['Shaadi','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.set(rc={'figure.figsize':(6,5)})
sns.barplot(x='Shaadi',y='Amount',data=sales_state,hue='Gender')
plt.show()
```

```python
# married woman purchased more
```

```python
# OCCUPATION
```

```python
ax=sns.countplot(x='Occupation',data=df)
sns.set(rc={'figure.figsize':(20,5)})

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
occu_sales=df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.set(rc={'figure.figsize':(10,5)})
sns.barplot(x='Occupation',y='Amount',data=occu_sales)
plt.show()
```

```python
# it sectors employees spend more and have the most  purchasing power than anybody else
```

```python
# PRODUCT CATEGORY
```

```python
df.columns
```

```python
ax=sns.countplot(x='Product_Category',data=df)
sns.set(rc={'figure.figsize':(24,6)})

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
pc_sales=df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(5)
sns.set(rc={'figure.figsize':(10,5)})
sns.barplot(x='Product_Category',y='Amount',data=pc_sales)
plt.show()
```

```python
# most sold product is food and most ordered product is clothing and apparels.
```

```python
product_id_sales=df.groupby(['Product_ID'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(5)
sns.set(rc={'figure.figsize':(10,5)})
sns.barplot(x='Product_ID',y='Amount',data=product_id_sales)
plt.show()
```

```python
# from the above analysis we can see that married woman age group of 26-35 from
# UP,MAHARASTRA,KARNATAKA working in IT,AVIATION AND HEALTHCARE 
# are more likely to buy food,cloting and apparel,electronic gadgets.
```
